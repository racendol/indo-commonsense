# indograd

```python
train_indograd = pd.read_csv('indograd/indograd_train.csv')
train_indograd
```

```python
val_indograd = pd.read_csv('indograd/indograd_val.csv')
val_indograd
```

```python
test_indograd = pd.read_csv('indograd/indograd_test.csv')
test_indograd
```

```python
test_indograd_dct = test_indograd.to_dict()
```

```python
answers_lst = []

for i in range(len(test_indograd)):
    label = test_indograd_dct['label'][i]
    if label == test_indograd_dct['option1'][i]:
        answers_lst.append(1)
    else:
        answers_lst.append(2)
```

```python
test_indograd['answer'] = answers_lst
```

```python
test_indograd
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('xlm-roberta-large')
```

```python
train_indograd_tokenized = tokenize_g(train_indograd, tokenizer, max_len=256, data_type='indograd')
val_indograd_tokenized = tokenize_g(val_indograd, tokenizer, max_len=256, data_type='indograd')
test_indograd_tokenized = tokenize_g(test_indograd, tokenizer, max_len=256, data_type='winogrande')
```

```python
y_train_indograd = tf.one_hot(train_indograd_tokenized[-1], depth=2)
y_val_indograd = tf.one_hot(val_indograd_tokenized[-1], depth=2)
y_test_indograd = tf.one_hot(test_indograd_tokenized[-1], depth=2)
```

```python
model = build_classifier_model_wsc_hf('xlm-roberta-large', seq_len=256, from_pt=True)

loss = tf.keras.losses.CategoricalCrossentropy(from_logits=True)
metrics = tf.metrics.CategoricalAccuracy()

epochs = 5
batch_size = 4

steps_per_epoch = int(len(train_indograd_tokenized) / batch_size)
num_train_steps = steps_per_epoch * epochs
num_warmup_steps = int(num_train_steps * 0.1)

init_lr = 1e-5
optimizer = optimization.create_optimizer(init_lr=init_lr,
                                          num_train_steps=num_train_steps,
                                          num_warmup_steps=num_warmup_steps,
                                          optimizer_type='adamw')

model.compile(optimizer=optimizer,
                         loss=loss,
                         metrics=metrics)
```

```python
callbacks = [
        tf.keras.callbacks.ModelCheckpoint('model_checkpoint.h5', verbose=1, save_best_only=True, monitor='val_categorical_acc')
]
```

```python
def train_model(x, y, model, epochs, batch_size, callbacks, val_x=None, val_y=None):
    val_data = None
    if val_x is not None and val_y is not None:
        val_x = val_x[:-1]
        
        val_data = (val_x, val_y)
    
    x = x[:-1]
    
    model.fit(
        x=x,
        y=y,
        batch_size=batch_size,
        epochs=1,
        verbose='auto',
        callbacks=callbacks,
        validation_data=val_data,
    )

def eval_model(x, y, model):
    loss, accuracy = model.evaluate(x[:-1], y)

    print(f'Loss: {loss}')
    print(f'Accuracy: {accuracy}')
    
```

```python
train_model(train_indograd_tokenized, y_train_indograd, model, epochs, batch_size, callbacks, val_indograd_tokenized, y_val_indograd)
```

```python
eval_model(test_indograd_tokenized, y_test_indograd, model)
```
